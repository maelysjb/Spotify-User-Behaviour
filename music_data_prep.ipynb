## Read Data
### I. EchoNest
Type of Data File: TXT

* Taste Profile User
* Train Triplets

```python
# Taste profile User
# The read_csv technique DOES NOT work: Error Tokenizing Data 

import csv

# file path and name for the taste profile user data
file_echo = '/EchoNest/taste_profile_usercat_120k.txt'

# reading the txt file through a csv file reader
taste_txt = open(path_init+file_echo,'rt')
reader = csv.reader(taste_txt)

# once contents are available, put them in a list
csv_list = []
for row in reader:
    # create temporary list and reformat for consistency 
    # remove beginning "### " and remove middle arrow " --> "
    temp_row = row[0].strip("# ").split(" --> ")
    # append to main list
    csv_list.append(temp_row)
taste_txt.close()

# convert list to pandas df (iloc to remove the extra None column at the end)
taste_profile = pd.DataFrame(csv_list, columns = ["Catalog Name", "Catalog ID", "None"]).iloc[:, :-1]
taste_profile.tail()
```

### IV. SQL lite Million Song Challenge
Type of Data File: DB (SQL Data)
https://datacarpentry.org/python-ecology-lesson/09-working-with-sql/index.html

```python
import sqlite3

# Similarity Table from artist similarity (db file)

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect(path_init + "/SQLlite_MillionSong/artist_similarity.db")
similarity_df = pd.read_sql_query("SELECT * from similarity", con)

# Verify that result of SQL query is stored in the dataframe
similarity_df.head()

con.close()
```

```python
similarity_df.describe()
```

```python
# make networkx with all similar artists
import networkx as nx
G = nx.Graph()
G = nx.from_pandas_edgelist(similarity_df, 'target', 'similar')
```
